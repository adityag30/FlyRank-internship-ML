## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

This project is primarily a **ranking and scoring** problem. The objective is to rank content pages based on their likelihood of needing a review or refresh rather than simply classifying them into categories. Each page receives a score calculated from observable search and engagement signals such as impressions, CTR, average position, content freshness, and user engagement.

The ranked output helps content managers and SEO analysts decide which pages should be reviewed first when time and resources are limited. Although a classification target can be used during model training (such as identifying declining pages), the final goal is to produce an evidence-based ranked review queue that supports human decision-making.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The target for this project is a **proxy label** called **is_declining_label**. It identifies whether a content page is showing a declining trend based on the trend_direction column in the starter dataset.

The proxy is defined as:

- **1 (Positive):** trend_direction = "down"
- **0 (Negative):** Any other trend direction

This label is based on a **defined rule** rather than a future observed outcome. It is useful for learning the machine learning workflow in the starter dataset, but it is not the final business target. In a larger warehouse dataset, a stronger target would be created using future performance windows, such as predicting whether a page will lose traffic over the next 30 days.

```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Create the proxy target
df["is_declining_label"] = (df["trend_direction"] == "down").astype(int)

print("Sample of the target column:")
display(df[["trend_direction", "is_declining_label"]].head(10))

print("\nTarget Distribution:")
print(df["is_declining_label"].value_counts())
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The primary success metric for this task is **Precision@50**.

Since the goal is to rank pages for review, Precision@50 measures how many of the top 50 recommended pages actually match the target (declining pages). A higher Precision@50 means reviewers spend more time on pages that truly need attention and less time on pages that do not.

For this task, a **higher Precision@50 is better**. Based on the starter notebook, a Precision@50 of around **0.68 or higher** indicates strong performance for the ranking model.

```python
# This cell is for CODE (numbers, a query, a check).

print("Target Distribution:")
print(df["is_declining_label"].value_counts())

print("\nPositive Rate:")
print(df["is_declining_label"].mean())
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule can only check a small number of conditions, such as "if impressions are high and CTR is low, then review the page." However, content performance depends on many observable signals that interact with each other, including impressions, clicks, CTR, average position, content freshness, sessions, and trend direction.

Machine learning can learn these patterns from data and combine multiple signals to produce a ranked list of pages that are most likely to benefit from a review. This makes the recommendations more flexible and data-driven than a single if-statement.

The model provides **decision-support** for content managers by prioritizing pages for review, while the final decision remains with a human reviewer.
