## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

This project improves the decision of **which pages should be prioritized for review**. Since content teams have limited time and resources, they need an evidence-based way to decide which pages are most likely to benefit from a refresh.<br>
The recommendations are intended for **content managers, SEO analysts, or content editors**. Based on the ranked list, they can refresh outdated content, improve metadata, expand thin content, optimize user engagement, or monitor pages showing early signs of decline.<br>
A **false positive** means reviewers spend time on pages that did not actually need attention, wasting valuable resources.<br>
A **false negative** means an important page that is losing visibility is overlooked, potentially resulting in reduced search traffic and missed business opportunities.<br>
Because review capacity is limited, producing an accurate ranked list of pages is more valuable than simply predicting whether a page is declining.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

This project uses **observed and measured search and engagement signals** to rank pages that may deserve review. My findings are **directional** and are intended to support decision-making rather than replace human judgment.

The recommendations produced by this project are **decision-support** tools. They help content managers and SEO analysts identify pages that appear to have refresh opportunities based on the available data.

This project **cannot prove** that refreshing a page will cause its search performance to improve. It also **cannot predict Google's search algorithm or ranking decisions**, nor can it guarantee that a recommended page will recover after being updated. Any recommendations should be reviewed by humans before action is taken.

Throughout this project, I will use careful language such as **"observed," "measured," "associated with," "suggests,"** and **"decision-support"** instead of making causal or guaranteed claims.
